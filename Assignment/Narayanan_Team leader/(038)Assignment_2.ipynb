# 6. Find the outliers and replace the outliers

```python
sns.boxplot(x=data['Gender']) # creating a boxplot of the age
# the points in the right side are the outliers
```

```python
q1 = data['Gender'].quantile(0.25)
q3 = data['Gender'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
data.loc[data['Gender'] < q1 - 1.5*iqr, 'Gender'] = q1
```

```python
data.loc[data['Gender'] > q3 + 1.5*iqr, 'Gender'] = q3
```

```python
sns.boxplot(x=data['Gender'])
```

# 7. check for categorical columns and perform encoding

```python
data.head()
```

```python
data["Geography"].unique()
```

```python
surnames = data["Surname"].unique()
surname_encoding = {}
for i in range(len(surnames)):
  surname_encoding[surnames[i]] = i
len(surname_encoding)
```

```python
# the categorial columns are gender and geography and surname
# encoding
data["Geography"].replace({"France":0,"Spain":1,"Germany":3}, inplace=True)
data["Gender"].replace({"Female":0,"Male":1}, inplace=True)
data["Surname"].replace(surname_encoding, inplace=True)
```

```python
data.head()
```
